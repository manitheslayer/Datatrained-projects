# creating categorical data dataframe

```python
dfd = pd.DataFrame()
dfd["Occupation"] = df["Occupation"]
dfd["Workclass"] = df["Workclass"]
dfd["Income"] = df["Income"]
dfd["Education_num"] = df['Education_num']
dfc = dfd.join(ordi)
dfc.shape
```

# removing the outlier removed rows in categorical dataframe

```python
dfc.drop(outlier[0],axis=0,inplace=True)
```

```python
dfc.shape
```

```python
dft = df1.join(dfc)
dft.drop("Income",axis=1,inplace=True)
dft.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(dft)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
xi = np.arange(1, 69, step=1)
yi = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, yi, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 68, step=1)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=1, color='r', linestyle='-')
plt.text(0.5, 0.85, '100% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

# observation

we can see from the plot 19 columns is more than enought to explain the variance and all information is retained within them

```python
pca=PCA(n_components=19)
dft=pca.fit_transform(dft)
dft=pd.DataFrame(dft)
dft.head()
```

# Importing required libraries

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
```

# i have write a code block that tests data with models and rank them according to cross val score and accuracy score

```python
algo = [DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),AdaBoostClassifier(),LogisticRegression()]
result = pd.DataFrame(columns=["Algorithm Name","Accuracy Score","Cross Validation Score"])

dtc=[]
knc=[]
rfc=[]
abc=[]
lr=[]
fl = [dtc,knc,rfc,abc,lr]


oo=0
for i in algo:
    rand=0
    acc=0
    for ii in range(20,60):
        train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.20,random_state=ii)
        cf = i
        cf.fit(train_x,train_y)
        pred = cf.predict(test_x)
        ac = accuracy_score(test_y,pred)
        if ac>acc:
            acc=ac
            rand=ii
    print(f' the best random state is {rand} and accuracy score is {acc} for algorithm {i}')
    
    
    train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.20,random_state=rand)
    cf = i
    cf.fit(train_x,train_y)
    pred = cf.predict(test_x)
    ac = accuracy_score(test_y,pred)
    cv = cross_val_score(cf,x_final,y,cv=5).mean()
    fl[oo].insert(0,i)
    fl[oo].insert(1,ac)
    fl[oo].insert(2,cv)
    result.loc[oo] = fl[oo]
    oo +=1
    
    
final_result = result.sort_values(by=["Accuracy Score","Cross Validation Score"],ascending=False)    
```

# HyperParameter Tunining

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier()
parameters = {'bootstrap': [True, False],
 'max_depth': [10, None],
 'max_features': ['auto', 'sqrt'],
 'min_samples_split': [2, 5],
 'n_estimators': [200,300]}
train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.20,random_state=47)
gsv = GridSearchCV(rf,parameters)
gsv.fit(train_x,train_y)
```

# using the best parameters got in grid search 

```python
rf = RandomForestClassifier(bootstrap=False,max_features='auto',n_estimators=200,min_samples_split=2,max_depth=None)
train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.20,random_state=47)
rf.fit(train_x,train_y)
pred = rf.predict(test_x)
ac = accuracy_score(test_y,pred)
cv = cross_val_score(rf,x_final,y,cv=5).mean()
print(f' the accuracy socre is {ac}, the cross validation score is {cv} ' )
```
