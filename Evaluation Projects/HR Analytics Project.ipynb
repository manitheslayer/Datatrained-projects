# Model Training Starts

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
```

# This Code block is written by me for calculating the best model and ranks them 

```python
algo = [LogisticRegression(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),AdaBoostClassifier(),SVC()]
result = pd.DataFrame(columns=["Algorithm Name","Accuracy Score","Cross Validation Score"])

lg =[]
dtc=[]
knc=[]
rfc=[]
abc=[]
svc=[]
fl = [lg,dtc,knc,rfc,abc,svc]

rand=0
acc=0
oo=0
for i in algo:
    for ii in range(0,100):
        train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=.20,random_state=ii)
        cf = i
        cf.fit(train_x,train_y)
        pred = cf.predict(test_x)
        ac = accuracy_score(test_y,pred)
        if ac>acc:
            acc=ac
            rand=ii
    print(f' the best random state is {rand} and accuracy score is {acc} for algorithm {i}')
    
    
    train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=.20,random_state=rand)
    cf = i
    cf.fit(train_x,train_y)
    pred = cf.predict(test_x)
    ac = accuracy_score(test_y,pred)
    cv = cross_val_score(cf,x,y,cv=5).mean()
    fl[oo].insert(0,i)
    fl[oo].insert(1,ac)
    fl[oo].insert(2,cv)
    result.loc[oo] = fl[oo]
    oo +=1
    
    
final_result = result.sort_values(by=["Accuracy Score","Cross Validation Score"],ascending=False)  
```

# HyperParamerter Tuning

```python
parameters = {'bootstrap': [True, False],
 'max_depth': [10, 20,50, 90, None],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [100,200, 400]}
```

```python
rf = RandomForestClassifier()
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=.20,random_state=39)
from sklearn.model_selection import GridSearchCV
gsv = GridSearchCV(rf,parameters)
gsv.fit(train_x,train_y)
```

# Running the Random Forest with best Parameters

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=39,test_size=0.2)
rf = RandomForestClassifier(bootstrap = False,
 max_depth = 90,
 max_features='auto',
 min_samples_leaf= 1,
 min_samples_split= 2,
 n_estimators= 200)
rf.fit(train_x,train_y)
pred = rf.predict(test_x)
ac = accuracy_score(test_y,pred)
con = confusion_matrix(test_y,pred)
cr = classification_report(test_y,pred)
cvs = cross_val_score(rf,x,y,cv=10).mean()
print( f' the accuracy score is {ac}')
print(con)
print(cr)
print(f' the cross val score is {cvs}')
```
