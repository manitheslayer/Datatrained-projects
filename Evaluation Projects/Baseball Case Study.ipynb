# observation

we can most of the columns are skewed so we are proceeding for skeweness removal using power transform

# using power transform makes R ,AB , H column to zero so i am going to use other methods for them 

after using log,sqrt,cbrt i conclude log works best

```python
x1 = pd.DataFrame()
x1["R"] = df["R"]
x1["AB"] = df["AB"]
x1["H"] = df["H"]
x1.skew()
```

```python
li = ["R","AB","H"]
for i in li:
    x1[i] = np.log(x1[i])   
```

```python
x1.skew()
```

```python
x.drop(["R","AB","H"],axis=1,inplace=True)
from sklearn.preprocessing import power_transform
x_new = power_transform(x)
x_n = pd.DataFrame(x_new,columns=x.columns)
x_n.skew()
```

```python
x_n["R"] = x1["R"]
x_n["AB"] = x1["AB"]
x_n["H"]  = x1["H"]
```

# Model training starts

# importing the Required Libraries for Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.svm import SVR
```

# This Code block is written by me for calculating the best model and ranks them 

```python
result = pd.DataFrame(columns=["Model Name","Train Score","Test Score","Cross Val Score"])
lin=[]
dec=[]
rid=[]
las=[]
kne=[]
rfr=[]
ada=[]
licol=[lin,dec,rid,las,kne,rfr,ada]

algo = [LinearRegression(),DecisionTreeRegressor(),Ridge(),Lasso(),KNeighborsRegressor(),RandomForestRegressor(),AdaBoostRegressor()]
oo= 0    
for v in algo:
    r = 0
    acc = 0
    for i in range(0,50):
        al = v
        train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.15,random_state=i)
        al.fit(train_x,train_y)
        score = al.score(train_x,train_y)
        if score>acc:
            acc = score
            r = i
            
    print(f'the best random state is {r} and best accuracy is { acc} for {v}')     
        
 
    train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.20,random_state=r)
    al.fit(train_x,train_y)
    trs = al.score(train_x,train_y)
    tss = al.score(test_x,test_y)
    cvs = cross_val_score(al,x_final,y).mean()
    print(f'the training score is {trs} the testing score is {tss} the cross val score is {cvs} for {v}')    
    licol[oo].insert(0,v)
    licol[oo].insert(1,trs)
    licol[oo].insert(2,tss)
    licol[oo].insert(3,cvs)
    result.loc[oo] = licol[oo]
    oo+=1

final_result = result.sort_values(by=["Cross Val Score","Test Score"],ascending=False)   
```

```python
final_result
```

# Observation

# Lasso performed best so proceeding with Lasso model for HyperParameter Tuning

# HyperParameter Tuning

```python
ls =Lasso()
train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.20,random_state=9)
parameters = {"alpha":[0.0001,0.001,0.01,0.05,0.1,0.2,0.5,0.8,1],"fit_intercept":[True,False],"copy_X":[True,False],"tol":[0.0001,0.001,0.1,0.5],"warm_start":[True,False]}
from sklearn.model_selection import GridSearchCV
gsv = GridSearchCV(ls,parameters)
gsv.fit(train_x,train_y)
```

```python
gsv.best_estimator_
```

```python
gsv.best_params_
```

# Running the Lasso Again with best Parameters

```python
ls =Lasso(alpha=1,tol=0.001,warm_start=True,copy_X=True,fit_intercept=True)
train_x,test_x,train_y,test_y = train_test_split(x_final,y,test_size=.15,random_state=9)
ls.fit(train_x,train_y)
trs = ls.score(train_x,train_y)
tss = ls.score(test_x,test_y)
pred = ls.predict(test_x)
cvs = cross_val_score(ls,x_final,y).mean()
print(f'the training score is {trs} the testing score is {tss} the cross val score is {cvs}')
print("Mean Squared Error",mean_squared_error(test_y,pred))
print("Mean Absolute Error", mean_absolute_error(test_y,pred))
print("Root Mean Squared Error", np.sqrt(mean_squared_error(test_y,pred)))
print("R2 Score", r2_score(test_y,pred))
```
