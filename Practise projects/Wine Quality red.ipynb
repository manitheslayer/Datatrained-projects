# outlier removal

```python
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df2))
```

```python
df3 = df2[(z<3).all(axis=1)]
```

# balancing the data using smote

here we can clearly see quality is not balanced

```python
sns.countplot(df3.iloc[:,-1])
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
dx = df3.iloc[:,:-1]
dy = df3.iloc[:,-1]
x1,y = sm.fit_resample(dx,dy)
```

```python
sns.countplot(y)
```

Now data is balanced

# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sr = StandardScaler()
x = sr.fit_transform(x1)
```

# Model Traning starts

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

# finding the best random State

```python
maxAccu=0  # maximum accuracy
maxRs=0 # best random state value for this max accuracy is achieved

for i in range(1,200):
    x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=.20,random_state = i)
    dtc = DecisionTreeClassifier()
    dtc.fit(x_train,y_train)
    pred = dtc.predict(x_test)
    acc = accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRs = i
        
        
print(f'MaxAccuracy is {maxAccu} and for random state is {maxRs}' ) 
```

```python
# using the random state value we received as best
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=.20,random_state = 6)
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred = dtc.predict(x_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print("cross val score is ",cross_val_score(dtc,x,y,cv=5))
print("Mean cross val score is ",cross_val_score(dtc,x,y,cv=5).mean())
```
