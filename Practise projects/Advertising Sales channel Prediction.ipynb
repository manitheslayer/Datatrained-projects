# Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(df2.iloc[:,0:-1])
```

# Model Training Starts

# Linear Regression

```python
y = df2.iloc[:,-1]
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

# finding the best random state

```python
random = 0
rc = 0

for i in range(0,200):
    lr = LinearRegression()
    train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=i,test_size=.20)
    lr.fit(train_x,train_y)
    acc = lr.score(train_x,train_y)
    if acc> rc :
        random = i
        rc = acc
    
print(f'the best random state is {random}, the best accuracy is {rc}')    
```

# using the best random state

```python
lr = LinearRegression()
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=10,test_size=.20)
lr.fit(train_x,train_y)
acc = lr.score(train_x,train_y)
acc1 = lr.score(test_x,test_y)
cr = cross_val_score(lr,x,y).mean()
print(f'the accuracy score for training is {acc}, the accuracy score of testing {acc1}, crossval score {cr}')
```

# hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestRegressor()
```

```python
parameters = {'bootstrap': [True, False],
 'max_depth': [10, 20,50, 90, None],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [100,200, 400]}
```

```python
gsv = GridSearchCV(rf,parameters)
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=17,test_size=.20)
gsv.fit(train_x,train_y)
```

```python
gsv.best_params_
```

```python
gsv.best_score_
```

# using the best parameters for the model

```python
rf = RandomForestRegressor(bootstrap=True,max_depth=90,max_features="auto",min_samples_leaf=1, min_samples_split=2,n_estimators=200)
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=17,test_size=.20)
rf.fit(train_x,train_y)
acc =rf.score(train_x,train_y)
acc1 = rf.score(test_x,test_y)
cr = cross_val_score(rf,x,y).mean()
print(f'the accuracy score for training is {acc}, the accuracy score of testing {acc1}, crossval score {cr}')
```
