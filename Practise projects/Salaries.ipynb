# With removing outliers and without removing skewness

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x = df2.iloc[:,0:5]
x = sc.fit_transform(x)
y = df2.iloc[:,-1]
```

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=42,test_size=.20)
```

```python
lr = LinearRegression()
lr.fit(train_x,train_y)
print("train score",lr.score(train_x,train_y))
print("test score",lr.score(test_x,test_y))
score = cross_val_score(lr,x,y,cv=10)
print("cross Validation score is ",score.mean())
```

# Tuning the Hyper parameters

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
from sklearn.model_selection import GridSearchCV
ls = Lasso()
lr = Ridge()
le = ElasticNet()

es = [ls,lr,le]

p = {"alpha":[0.001,0.01,0.1,1,10]}

for i in es:
    gs = GridSearchCV(estimator=i,param_grid=p)
    gs.fit(train_x,train_y)
    print(gs.best_estimator_)
```

```python
ls = Lasso(alpha=10)
ls.fit(train_x,train_y)
print("train score",ls.score(train_x,train_y))
print("test score",ls.score(test_x,test_y))
score = cross_val_score(ls,x,y,cv=10)
print("cross Validation score is ",score.mean())
```

```python
ls = Ridge(alpha=10)
ls.fit(train_x,train_y)
print("train score",ls.score(train_x,train_y))
print("test score",ls.score(test_x,test_y))
score = cross_val_score(ls,x,y,cv=10)
print("cross Validation score is ",score.mean())
```

```python
ls = ElasticNet(alpha=0.1)
ls.fit(train_x,train_y)
print("train score",ls.score(train_x,train_y))
print("test score",ls.score(test_x,test_y))
score = cross_val_score(ls,x,y,cv=10)
print("cross Validation score is ",score.mean())
```

# Elastic net gives best accuracy overall
