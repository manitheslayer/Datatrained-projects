# Standard Scaler

```python
s = df_new.drop("Survived",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
sc = ss.fit_transform(s)
sc_final = pd.DataFrame(sc,columns=s.columns)
sc_final.shape
```

# Model Training starts

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
x = sc_final
y = df_new.iloc[:,5]
```

# Logisitc Regression

```python
from sklearn.linear_model import LogisticRegression
```

# finding the best random state

```python
rs = 0
acc = 0

for i in range(0,200):
    train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=i,test_size=0.2)
    lg = LogisticRegression()
    lg.fit(train_x,train_y)
    pred = lg.predict(test_x)
    ac = accuracy_score(test_y,pred)
    if ac > acc:
        acc = ac
        rs = i
print(f'the best random state is {rs} and the best accuracy is {acc}')        
```

# using the best random state

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=33,test_size=0.2)
lg = LogisticRegression()
lg.fit(train_x,train_y)
pred = lg.predict(test_x)
ac = accuracy_score(test_y,pred)
con = confusion_matrix(test_y,pred)
cr = classification_report(test_y,pred)
cvs = cross_val_score(lg,x,y,cv=10).mean()
print( f' the accuracy score is {ac}')
print(con)
print(cr)
print(f' the cross val score is {cvs}')
```

# hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier()
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=11,test_size=0.2)
```

```python
parameters = {'bootstrap': [True, False],
 'max_depth': [10, 20,50, 90, None],
 'max_features': ['auto', 'sqrt'],
 'min_samples_leaf': [1, 2, 4],
 'min_samples_split': [2, 5, 10],
 'n_estimators': [100,200, 400]}
```

```python
gsv = GridSearchCV(rf,parameters)
gsv.fit(train_x,train_y)
```

```python
gsv.best_estimator_
```

```python
gsv.best_params_
```

```python
gsv.best_score_
```

# using the best params we got from hyper parameter tuning

```python
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=11,test_size=0.2)
rf = RandomForestClassifier(bootstrap = True,
 max_depth = 20,
 max_features='auto',
 min_samples_leaf= 2,
 min_samples_split= 2,
 n_estimators= 200)
rf.fit(train_x,train_y)
pred = rf.predict(test_x)
ac = accuracy_score(test_y,pred)
con = confusion_matrix(test_y,pred)
cr = classification_report(test_y,pred)
cvs = cross_val_score(rf,x,y,cv=10).mean()
print( f' the accuracy score is {ac}')
print(con)
print(cr)
print(f' the cross val score is {cvs}')
```
