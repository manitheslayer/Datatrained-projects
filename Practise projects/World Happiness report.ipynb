```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()
```

# outlier removal

```python
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df1))
df2 = df1[(z<3).all(axis=1)]
df2
```

# without removing outlier and without removing skewness

```python
x = df1.drop("Happiness Score",axis=1)
y = df1.iloc[:,3]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
train_x,test_x,train_y,test_y = train_test_split(x,y,random_state=42,test_size=0.2)
lr.fit(train_x,train_y)
print("train Score",lr.score(train_x,train_y))
print("test score",lr.score(test_x,test_y))
c = cross_val_score(lr,x,y,cv=50)
print("cross val score",c.mean())
```
